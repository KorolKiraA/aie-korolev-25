```python
# Импорт необходимых библиотек с общепринятым сокращением
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("1. ИМПОРТ БИБЛИОТЕК")
print("Библиотеки импортированы успешно!")
```

```python
print("\n2. ЗАГРУЗКА ДАННЫХ")

# Загрузка датасета
df = pd.read_csv('S02-hw-dataset.csv')
print("Датасет загружен успешно!")

# Показываем первые строки
print("\nПервые 5 строк датасета:")
print(df.head())
```

```python
print("\n3. АТРИБУТЫ ДАТАФРЕЙМА")

# Форма датафрейма
print(f"Форма датасета (строки, столбцы): {df.shape}")

# Информация о типах данных
print("\nИнформация о типах данных:")
print(df.info())

# Имена столбцов
print(f"\nИмена столбцов: {list(df.columns)}")
```

```python
print("\n4. ОПИСАТЕЛЬНЫЕ СТАТИСТИКИ")

# Базовые статистики
print("Базовые статистики числовых столбцов:")
print(df.describe())

# Статистики для каждого столбца отдельно
print("\nСредние значения по столбцам:")
print(df.mean(numeric_only=True))
```

```python
print("\n5. ПРОВЕРКА КАЧЕСТВА ДАННЫХ")

# Проверка пропусков
print("Количество пропусков в каждом столбце:")
print(df.isna().sum())

# Проверка дубликатов
print(f"\nКоличество полных дубликатов строк: {df.duplicated().sum()}")

# Удаление дубликатов
df_clean = df.drop_duplicates()
print(f"После удаления дубликатов: {df_clean.shape[0]} строк")
```

```python
print("\n6. ФИЛЬТРАЦИЯ И ВЫБОРКА ДАННЫХ")

# Выбор столбцов
print("Столбцы 'age' и 'country':")
print(df[['age', 'country']].head())

# Фильтрация по условию
print("\nПользователи из России:")
russian_users = df[df['country'] == 'RU']
print(russian_users.head())

print("\nПользователи старше 30 лет:")
adult_users = df[df['age'] > 30]
print(adult_users.head())
```

```python
print("\n7. ГРУППИРОВКА И АГРЕГАЦИЯ")

# Группировка по странам
print("Средний возраст по странам:")
age_by_country = df.groupby('country')['age'].mean()
print(age_by_country)

print("\nОбщая выручка по странам:")
revenue_by_country = df.groupby('country')['revenue'].sum()
print(revenue_by_country)

print("\nКоличество пользователей по странам:")
users_by_country = df.groupby('country')['user_id'].count()
print(users_by_country)
```

```python
print("\n8. ВИЗУАЛИЗАЦИЯ ДАННЫХ")

# Создаем папку для графиков, если её нет
import os
if not os.path.exists('figures'):
    os.makedirs('figures')

# ГИСТОГРАММА распределения возраста
print("\n1. Гистограмма распределения возраста")
plt.figure(figsize=(8, 5))
plt.hist(df['age'].dropna(), bins=10, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Распределение возраста пользователей', fontsize=14)
plt.xlabel('Возраст', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figures/age_histogram.png', dpi=300, bbox_inches='tight')
plt.show()

# BOXPLOT выручки по странам
print("\n2. Boxplot выручки по странам")
plt.figure(figsize=(10, 6))
df.boxplot(column='revenue', by='country', grid=False)
plt.title('Распределение выручки по странам', fontsize=14)
plt.suptitle('')  # Убираем автоматический заголовок
plt.xlabel('Страна', fontsize=12)
plt.ylabel('Выручка', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figures/revenue_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()

# SCATTER PLOT: возраст vs выручка
print("\n3. Scatter plot: возраст vs выручка")
plt.figure(figsize=(8, 5))
plt.scatter(df['age'], df['revenue'], alpha=0.6, s=50, color='green')
plt.title('Зависимость выручки от возраста', fontsize=14)
plt.xlabel('Возраст', fontsize=12)
plt.ylabel('Выручка', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figures/age_vs_revenue_scatter.png', dpi=300, bbox_inches='tight')
plt.show()

# Дополнительный график: столбчатая диаграмма
print("\n4. Столбчатая диаграмма: средняя выручка по странам")
plt.figure(figsize=(8, 5))
revenue_by_country.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Средняя выручка по странам', fontsize=14)
plt.xlabel('Страна', fontsize=12)
plt.ylabel('Средняя выручка', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('figures/avg_revenue_bar.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("\n9. АНАЛИЗ ВЫБРОСОВ")

# Поиск выбросов в возрасте
print("Пользователи с подозрительным возрастом:")
age_outliers = df[(df['age'] < 0) | (df['age'] > 100)]
print(age_outliers[['user_id', 'age', 'country']])

# Поиск выбросов в покупках
print("\nПользователи с отрицательными покупками:")
negative_purchases = df[df['purchases'] < 0]
print(negative_purchases[['user_id', 'purchases', 'revenue']])

# Поиск нулевой выручки при наличии покупок
print("\nПользователи с покупками, но нулевой выручкой:")
zero_revenue = df[(df['purchases'] > 0) & (df['revenue'] == 0)]
print(zero_revenue[['user_id', 'purchases', 'revenue', 'country']])
```

```python
print("\n10. СОЗДАНИЕ НОВЫХ ПРИЗНАКОВ")

# Создаем признак "средний чек" с проверкой деления на ноль
df['avg_check'] = df.apply(
    lambda row: row['revenue'] / row['purchases'] if row['purchases'] > 0 else 0, 
    axis=1
)
print("Добавлен новый признак 'avg_check' (средний чек)")

# Создаем возрастные группы
bins = [0, 20, 30, 40, 50, 100]
labels = ['<20', '20-30', '30-40', '40-50', '50+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

print("Добавлен новый признак 'age_group' (возрастная группа)")

# Показываем статистики по возрастным группам
print("\nСредняя выручка по возрастным группам:")
# Используем observed=True чтобы избежать предупреждения
age_group_stats = df.groupby('age_group', observed=True)['revenue'].mean()
print(age_group_stats)

# Также можно вывести количество пользователей в каждой группе
print("\nКоличество пользователей по возрастным группам:")
age_group_counts = df.groupby('age_group', observed=True)['user_id'].count()
print(age_group_counts)

# Дополнительные статистики по возрастным группам
print("\nПолные статистики по возрастным группам:")
full_age_stats = df.groupby('age_group', observed=True).agg({
    'user_id': 'count',
    'age': 'mean',
    'purchases': 'mean',
    'revenue': ['mean', 'sum']
})
print(full_age_stats.round(2))
```

```python
print("\n11. КОРРЕЛЯЦИОННЫЙ АНАЛИЗ")

# Вычисляем корреляцию между числовыми признаками
numeric_columns = ['age', 'purchases', 'revenue', 'avg_check']
correlation_matrix = df[numeric_columns].corr()
print("Матрица корреляции:")
print(correlation_matrix)

# Визуализация корреляционной матрицы
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(numeric_columns)), numeric_columns, rotation=45)
plt.yticks(range(len(numeric_columns)), numeric_columns)
plt.title('Матрица корреляции числовых признаков', fontsize=14)
for i in range(len(numeric_columns)):
    for j in range(len(numeric_columns)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}", 
                ha="center", va="center", color="white" if abs(correlation_matrix.iloc[i, j]) > 0.5 else "black")
plt.tight_layout()
plt.savefig('figures/correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("\n12. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ")

# Сохраняем очищенные данные
df_clean.to_csv('figures/cleaned_data.csv', index=False)
print("Очищенные данные сохранены в файл 'figures/cleaned_data.csv'")

# Сохраняем сводную таблицу
summary_stats = df.groupby('country').agg({
    'user_id': 'count',
    'age': 'mean',
    'purchases': 'mean',
    'revenue': 'sum'
}).round(2)
summary_stats.to_csv('figures/summary_statistics.csv')
print("Сводная статистика сохранена в файл 'figures/summary_statistics.csv'")
print("\nСводная статистика по странам:")
print(summary_stats)
```
